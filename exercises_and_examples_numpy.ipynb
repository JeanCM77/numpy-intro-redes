## 3.3 Funciones en Python

````python
def calculate_average(numbers):
    nums=list(numbers)
    return sum(nums)/len(nums)

def find_power(base, exponent=2):
    return base**exponent

print(calculate_average([1,2,3,4]))
print(find_power(3))
print(find_power(2,3))
````

## 3.4 Descenso por gradiente

````python
X = np.array([1,2,3,4,5],dtype=float)
y = np.array([3,6,8,11,13],dtype=float)
m = len(X)
X_b = np.c_[np.ones((m,1)), X]

learning_rate = 0.01
n_iterations = 1000
theta = np.random.randn(2)
history_theta = np.zeros((n_iterations,2))
history_loss = np.zeros(n_iterations)

for i in range(n_iterations):
    preds = X_b.dot(theta)
    errors = preds - y
    gradients = (2/m)*X_b.T.dot(errors)
    theta -= learning_rate*gradients
    history_theta[i] = theta
    history_loss[i] = (errors**2).mean()

theta
````

````python
plt.plot(history_theta[:,0],label='theta0')
plt.plot(history_theta[:,1],label='theta1')
plt.legend(); plt.grid(); plt.show()

plt.scatter(X,y)
x_line = np.linspace(1,5,100)
y_line = theta[0] + theta[1]*x_line
plt.plot(x_line,y_line)
plt.grid(); plt.show()
````
